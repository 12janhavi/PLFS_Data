```python
import pandas as pd

# Load the dataset
data = pd.read_csv("new_file_with_final_weight.csv")

# Apply filters for different categories
rural_male = data[(data['Sector'] == 1) & (data.iloc[:, 4] == 1)]
rural_female = data[(data['Sector'] == 1) & (data.iloc[:, 4] == 2)]
urban_male = data[(data['Sector'] == 2) & (data.iloc[:, 4] == 1)]
urban_female = data[(data['Sector'] == 2) & (data.iloc[:, 4] == 2)]

# Calculate LFPR for each category
lfpr_rural_male = ((rural_male['Earnings For Regular Salaried/Wage Activity'].notnull() | rural_male['Earnings For Self Employed'].notnull()) & (rural_male['Age'] >= 16)).sum() / (rural_male['Age'] >= 16).sum() * 100
lfpr_rural_female = ((rural_female['Earnings For Regular Salaried/Wage Activity'].notnull() | rural_female['Earnings For Self Employed'].notnull()) & (rural_female['Age'] >= 16)).sum() / (rural_female['Age'] >= 16).sum() * 100 if len(rural_female) > 0 else 0
lfpr_urban_male = ((urban_male['Earnings For Regular Salaried/Wage Activity'].notnull() | urban_male['Earnings For Self Employed'].notnull()) & (urban_male['Age'] >= 16)).sum() / (urban_male['Age'] >= 16).sum() * 100 if len(urban_male) > 0 else 0
lfpr_urban_female = ((urban_female['Earnings For Regular Salaried/Wage Activity'].notnull() | urban_female['Earnings For Self Employed'].notnull()) & (urban_female['Age'] >= 16)).sum() / (urban_female['Age'] >= 16).sum() * 100 if len(urban_female) > 0 else 0

# Combine urban males and females
urban_male_female = pd.concat([urban_male, urban_female])

# Combine rural males and females
rural_male_female = pd.concat([rural_male, rural_female])

# Calculate LFPR for combined categories
lfpr_urban_male_female = (((urban_male_female['Earnings For Regular Salaried/Wage Activity'].notnull() | urban_male_female['Earnings For Self Employed'].notnull()) & (urban_male_female['Age'] >= 16)).sum() / (urban_male_female['Age'] >= 16).sum()) * 100 if len(urban_male_female) > 0 else 0
lfpr_rural_male_female = (((rural_male_female['Earnings For Regular Salaried/Wage Activity'].notnull() | rural_male_female['Earnings For Self Employed'].notnull()) & (rural_male_female['Age'] >= 16)).sum() / (rural_male_female['Age'] >= 16).sum()) * 100 if len(rural_male_female) > 0 else 0

print("LFPR for Rural Male:", lfpr_rural_male)
print("LFPR for Rural Female:", lfpr_rural_female)
print("LFPR for Urban Male:", lfpr_urban_male)
print("LFPR for Urban Female:", lfpr_urban_female)
print("LFPR for Urban Male + Female:", lfpr_urban_male_female)
print("LFPR for Rural Male + Female:", lfpr_rural_male_female)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the labels for each employment status code
employment_status_labels = {
    11: 'Own account worker',
    12: 'Employer',
    21: 'Helper in HH enterprises',
    31: 'Regular salaried/wage employees',
    41: 'Casual wage labor in public works',
    42: 'Casual wage labor in MGNREG works',
    51: 'Casual wage labor in other type of work'
}

# Replace the employment status codes with their corresponding labels
new_file_with_final_weight['Employment Status'] = new_file_with_final_weight['Current Weekly Status (CWS)'].map(employment_status_labels)

# Calculate the percentage of each employment status category within each sex group
grouped_data = new_file_with_final_weight.groupby(['Sex', 'Employment Status']).size().reset_index(name='Count')
sex_totals = grouped_data.groupby('Sex')['Count'].transform('sum')
grouped_data['Percentage'] = (grouped_data['Count'] / sex_totals) * 100
# Pivot the data to have technical education levels as columns
pivot_data = grouped_data.pivot(index='Employment Status', columns='Sex', values='Percentage')

# Print the resulting DataFrame
print(pivot_data)


# Define the order of employment status categories
order = [
    'Own account worker',
    'Employer',
    'Helper in HH enterprises',
    'Regular salaried/wage employees',
    'Casual wage labor in public works',
    'Casual wage labor in MGNREG works',
    'Casual wage labor in other type of work'
]

# Create the count plot with hue='Sex'
plt.figure(figsize=(12, 10))
sns.barplot(x='Employment Status', y='Percentage', hue='Sex', data=grouped_data, order=order)
plt.title('Distribution of Employment Status by Sex')
plt.xlabel('Employment Status')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Update the legend labels
handles, labels = plt.gca().get_legend_handles_labels()
labels = ['Male', 'Female', 'Transgender']  # Update the legend labels
plt.legend(handles, labels, title='Sex')

# Set the y-axis limits to scale the percentage values
plt.ylim(0, 100)

# Add percentage text on each bar
for bar in plt.gca().patches:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width() / 2, height + 2, f'{height:.1f}%', ha='center', color='black', fontsize=10)

plt.show()
```
